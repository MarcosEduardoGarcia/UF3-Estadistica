#Soluciones

```python
import matplotlib.pyplot as plt

horas_estudio = [1,2,2,3,4,4,5,6,6,7,8,3,5,2,4]
calificaciones = [55,60,65,70,72,75,80,85,88,90,95,68,78,62,74]

# --- Scatter Plot ---
plt.figure(figsize=(6,4))
plt.scatter(horas_estudio, calificaciones)

plt.title("Horas de estudio vs Calificación")
plt.xlabel("Horas de estudio")
plt.ylabel("Calificación")
plt.grid(True)

plt.show()

# --- Histograma ---
plt.figure(figsize=(6,4))
plt.hist(calificaciones, bins=5)

plt.title("Distribución de Calificaciones")
plt.xlabel("Calificación")
plt.ylabel("Frecuencia")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

producto_A = [20, 22, 19, 24, 30, 28, 26]
producto_B = [15, 18, 14, 20, 21, 19, 17]
producto_C = [10, 12, 11, 13, 15, 14, 16]

# Totales
totales = [
    sum(producto_A),
    sum(producto_B),
    sum(producto_C)
]

nombres = ["Producto A", "Producto B", "Producto C"]

# --- Bar Plot ---
plt.figure(figsize=(6,4))
plt.bar(nombres, totales)

plt.title("Ventas Totales por Producto")
plt.xlabel("Producto")
plt.ylabel("Ventas Totales")

plt.show()

# --- Boxplot ---
plt.figure(figsize=(6,4))
plt.boxplot([producto_A, producto_B, producto_C],
            labels=nombres)

plt.title("Distribución de Ventas Semanales")

plt.show()
```

```python
import matplotlib.pyplot as plt

temperatura = [22, 24, 23, 25, 27, 26, 24]
dias = ["Lun","Mar","Mie","Jue","Vie","Sab","Dom"]

plt.figure(figsize=(12,4))

# --- Line Plot ---
plt.subplot(1,3,1)
plt.plot(dias, temperatura, marker='o')
plt.title("Temperatura diaria")
plt.xlabel("Día")
plt.ylabel("°C")
plt.grid(True)

# --- Histogram ---
plt.subplot(1,3,2)
plt.hist(temperatura, bins=4)
plt.title("Distribución")

# --- Boxplot ---
plt.subplot(1,3,3)
plt.boxplot(temperatura)
plt.title("Variabilidad")

plt.suptitle("Dashboard de Temperatura Semanal")

plt.tight_layout()
plt.show()
```
